```python
import numpy as np
import sympy as sy
from matplotlib import pyplot as plt
import math
```

```python
#############Problem 1#############
def f(x):
    return (sy.sin(x) + 1) ** (sy.sin(sy.cos(x)))
x = sy.symbols('x')

f_x = sy.lambdify(x, sy.diff(f(x), x))
#lambdify can make the f(x) accept np.array
x_array = np.linspace(-math.pi, math.pi)
fx = sy.lambdify(x, f(x))
ax = plt.gca()
ax.spines["bottom"].set_position("zero")
plt.plot(x_array, f_x(x_array), label = 'derivative')
plt.plot(x_array, fx(x_array), label = 'origin')
plt.legend()
plt.show()
```

```python
#############Problem 2#############

#Forward 1
def g_x(s, x, h):
    return (s(x + h) - s(x)) / h

#Forward 2
def q_x(s, x, h):
    return (-3 * s(x) + 4 * s(x + h) - s(x + 2 * h )) / (2 * h)

#Backward 1
def w_x(s, x, h):
    return (s(x) - s(x - h)) / h

#Backward 2
def h_x(s, x, h):
    return (3 * s(x) - 4 * s(x - h) + s(x - 2 * h )) / (2 * h)

#Centered 1
def j_x(s, x, h):
    return (s(x + h) - s(x - h)) / (2 * h)

#Centered 2
def k_x(s, x, h):
    return (s(x - 2 * h) - 8 * s(x - h) + 8 * s(x + h) - s(x + 2 * h )) / (12 * h)

plt.plot(x_array, f_x(x_array), label = 'derivative')
plt.plot(x_array, g_x(fx, x_array, 0.01), label = 'Forward 1', color = 'black')
plt.plot(x_array, q_x(fx, x_array, 0.01), color = 'red', label = 'Forward 2')
plt.plot(x_array, w_x(fx, x_array, 0.01), color = 'green', label = 'Backward 1')
plt.plot(x_array, h_x(fx, x_array, 0.01), color = 'brown', label = 'Backward 2')
plt.plot(x_array, g_x(fx, x_array, 0.01), label = 'Centered 1', color = 'purple')
plt.plot(x_array, k_x(fx, x_array, 0.01), color = 'pink', label = 'Centered 2')
ax = plt.gca()
ax.spines["bottom"].set_position("zero")
plt.legend()
plt.show()
```

```python
#############Problem 3#############

h_array = np.logspace(-8, 0, 9)
f_list = [g_x, q_x, w_x, h_x, j_x, k_x]
def absol_error(l, x, h):
    return abs(f_x(x) - l(fx, x, h))
plt.figure()
for t in f_list:
    plt.plot(h_array, absol_error(t, 1, h_array), marker = 'o')
plt.loglog()
plt.legend(['Forward 1', 'Forward 2', 'Backward 1', 'Backward 2', 'Centered 1', 'Centered 2'], loc='lower right')
axes = plt.gca()
plt.show()
```

```python
#############Problem 4#############

import pandas as pd
data = np.load('plane.npy')

def x(a, b):
    alpha = np.deg2rad(a)
    beta = np.deg2rad(b)
    return 500 * sy.tan(beta) / (sy.tan(beta) - sy.tan(alpha))

def y(a, b):
    alpha = np.deg2rad(a)
    beta = np.deg2rad(b)
    return 500 * sy.tan(beta) * sy.tan(alpha) / (sy.tan(beta) - sy.tan(alpha))

a_b_array = data[:, 1:3]
xy = []
for i in a_b_array:
    x_cor = x(i[0], i[1])
    y_cor = y(i[0], i[1])
    xy.append((x_cor,y_cor))
    
def value_x(t):
    return xy[t-7][0]

def value_y(t):
    return xy[t-7][1]

def speed(x, y):
    return (x ** 2 + y ** 2) ** (1/2)

# t = 7
speed_7 = speed(g_x(value_x, 7, 1), g_x(value_y, 7, 1))

# t = 14
speed_14 = speed(w_x(value_x, 14, 1), w_x(value_y, 14, 1))

#t = 8, 9, ..., 13
speed = [speed(j_x(value_x, i, 1), j_x(value_y, i, 1)) for i in range(8, 14)]
speed.insert(0, speed_7)
speed.append(speed_14)
speed_array = np.array(speed)
speed_df = pd.DataFrame(data = speed_array, columns = {'speed'}, index = {7, 8, 9, 10, 11, 12, 13, 14})
speed_df
```

```python
#############Problem 5#############

def Japprox(f,vx,h = 0.1):
    dim_f = len(f)
    dim_x = len(vx)
    I = np.eye(dim_x)
    Jmatrix = np.zeros((dim_f, dim_x))
    for i, sub_f in enumerate(f):
        for j,s in enumerate(vx):
            func = sy.lambdify((x, y), sub_f)
            forward = vx + h*I[:,j]
            back = vx - h*I[:,j]
            Jmatrix[i][j] = (func(forward[0],forward[1])-func(back[0],back[1]))/(2.0*h)
    return Jmatrix
```

```python
x = sy.symbols('x')
y = sy.symbols('y')
f_test = [x ** 2, x ** 3 - y]
x_test = [5, 7]
print(f_test, x_test, Japprox(f_test,x_test))
```

```python
#############Problem 7#############

import time
from autograd import grad
import random
from autograd import numpy as anp
g1 = lambda x: (sy.sin(x) + 1) ** (sy.sin(sy.cos(x)))
g2 = lambda x: (anp.sin(x) + 1) ** anp.sin(anp.cos(x))
ope_time_1 = []
ope_time_2 = []
ope_time_3 = []
error_1 = []
error_2 = []
error_3 = []
N = 200
#Calculate exact value
for i in range(N):
    x_0 = random.uniform(-2*np.pi, 2*np.pi)
    start_1 = time.time()
    x = sy.symbols('x')   
    fx = sy.lambdify(x, sy.diff(g1(x), x))
    a = fx(x_0)
    end_1 = time.time()
    ope_time_1.append(end_1 - start_1)
    error_1.append(1e-18)

#Calculate fourth-order centered difference quotient
    start_2 = time.time()
    d = k_x(g1, x_0, 0.01)
    end_2 = time.time()
    ope_time_2.append(end_2 - start_2) 
    error_2.append(abs(a - d))
    
#Calculate Autograd
    start_3 = time.time()
    df = grad(g2)
    c = df(x_0)
    end_3 = time.time()
    ope_time_3.append(end_3 - start_3) 
    error_3.append(abs(a - c))

#draw the plot
plt.scatter(ope_time_1, error_1, alpha=0.4, label='SymPy')
plt.scatter(ope_time_2, error_2, alpha=0.4, label='Difference Quotients')
plt.scatter(ope_time_3, error_3, alpha=0.4, label='Autograd')
plt.yscale('log')
plt.xscale('log')
plt.xlim(1.0e-5,1.0e-2)
plt.ylim(1.0e-19,1.0e-6)
plt.xlabel('Computation Time (seconds)')
plt.ylabel('Absolute Error')
plt.legend()
plt.show()
```
